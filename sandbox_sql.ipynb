### JOIN

```python
query = "SELECT * FROM 'transaction'"

transaction = pd.read_sql(query, connection)
transaction
```

```python
query = "SELECT * FROM 'user'"

user = pd.read_sql(query, connection)
user
```

```python
query = "SELECT * FROM 'product'"

product = pd.read_sql(query, connection)
product
```

```python
query = """
    SELECT * 
    FROM 'transaction'
        LEFT JOIN 'user'
        ON 'transaction'.Name = 'user'.Name
"""

transaction = pd.read_sql(query, connection)
transaction
```

```python
query = """
    SELECT 'transaction'.Name, 'transaction'.result, 'user'.Age, 'user'.Sex
    FROM 'transaction'
        LEFT JOIN 'user'
        ON 'transaction'.Name = 'user'.Name
"""

transaction = pd.read_sql(query, connection)
transaction
```

```python
# 구매이력이 없는 고객까지 출력하고자 할 때는 'user'를 왼족에 JOIN 합니다.

query = """
    SELECT 'user'.Name, 'transaction'.result, 'user'.Age, 'user'.Sex
    FROM 'user'
        LEFT JOIN 'transaction'
        ON 'transaction'.Name = 'user'.Name
"""

transaction = pd.read_sql(query, connection)
transaction
```

```python
# 기본적으로는 INNER JOIN 은 하게 됩니다.

query = """
    SELECT 'user'.Name, 'transaction'.result, 'user'.Age, 'user'.Sex
    FROM 'transaction'
        JOIN 'user'
        ON 'transaction'.Name = 'user'.Name
"""

transaction = pd.read_sql(query, connection)
transaction
```
