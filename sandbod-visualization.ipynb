## distplot

```python
# bins == 막대의 갯수

sns.distplot(data["허리둘레"], bins=50)
```

```python
# hist == 히스토그램 유무

sns.distplot(data["허리둘레"], hist=False)
```

```python
# kde == 가우시안 그래프 유무

sns.distplot(data["허리둘레"], kde=False)
```

```python
data["흡연상태"].value_counts()
```

```python
one = data[data["흡연상태"] == 1.0]
two = data[data["흡연상태"] == 2.0]
three = data[data["흡연상태"] == 3.0]
```

```python
sns.distplot(one["허리둘레"], label = "흡연 안함", hist=False)
sns.distplot(two["허리둘레"], label = "흡연 했지만 그만둠", hist=False)
sns.distplot(three["허리둘레"], label = "흡연 하고 있음", hist=False)
```

## barplot

```python
plt.figure(figsize=(18,3))

# pd.np.mean ==> pandas에서 numpy의 평균을 불러와 사용합니다.
# 각 바의 세로 검은 막대는 편차입니다.

sns.barplot(data=data, x="연령대코드(5세단위)", y="수축기혈압", estimator=pd.np.mean)
```

```python
plt.figure(figsize=(18,3))

sns.barplot(data=data, x="연령대코드(5세단위)", y="수축기혈압", estimator=pd.np.sum)
```

```python
# x축, y축을 위 그래프와 바꿔 그립니다.

sns.barplot(data=data, y="연령대코드(5세단위)", x="수축기혈압")
```

```python
# orient == 막대 그래프의 방향 설정

plt.figure(figsize=(4,7))

sns.barplot(data=data, y="연령대코드(5세단위)", x="수축기혈압", orient='h')
```

## boxplot

```python
plt.figure(figsize=(18,3))

sns.boxplot(data=data, x="연령대코드(5세단위)", y="수축기혈압")

# 각 박스 아래 위의 가로줄은 밑에서 부터 최소값, 25%, 중간값, 75%, 최대값을 나타냅니다.
# x 축의 18이 가로줄 하나만 있는 이유는 값이 1개 라서 그러합니다.
# 점으로 표현된것은 아웃라이어로 볼수 있습니다.
```

```python
plt.figure(figsize=(18,3))

sns.boxplot(data=data, x="연령대코드(5세단위)", y="수축기혈압", hue="성별코드")
```

```python
plt.figure(figsize=(18,3))

sns.violinplot(data=data, x="연령대코드(5세단위)", y="수축기혈압", hue="성별코드")
```

## pointplot

```python
plt.figure(figsize=(18,3))

sns.pointplot(data=data, x="연령대코드(5세단위)", y="수축기혈압")
```

```python
# ci == 극단값 설정

plt.figure(figsize=(18,3))

sns.pointplot(data=data, x="연령대코드(5세단위)", y="수축기혈압", ci=99)
```

```python
plt.figure(figsize=(18,3))

sns.pointplot(data=data, x="연령대코드(5세단위)", y="수축기혈압", estimator=sum)
```

```python
plt.figure(figsize=(18,3))

sns.pointplot(data=data, x="연령대코드(5세단위)", y="수축기혈압", hue="성별코드")
```

```python
plt.figure(figsize=(18,3))

sns.pointplot(data=data, x="연령대코드(5세단위)", y="수축기혈압", linestyles='--', hue="성별코드")
```

```python
plt.figure(figsize=(18,3))

sns.pointplot(data=data, x="연령대코드(5세단위)", y="수축기혈압", linestyles=[':', '--'], hue="성별코드")
```
